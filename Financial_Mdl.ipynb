# **Financial Modelling**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Example synthetic data for the market (Yearly Revenue in billions USD)
data = {
    'Year': np.arange(2010, 2024),
    'Revenue': [50, 55, 60, 67, 73, 80, 85, 90, 97, 104, 110, 118, 125, 130]
}
```

```python
# Convert to DataFrame
df = pd.DataFrame(data)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Revenue'], marker='o')
plt.xlabel('Year')
plt.ylabel('Revenue (in billions USD)')
plt.title('Market Revenue Over Time')
plt.show()
```

```python
# Prepare the data for the regression model
X = df[['Year']]
y = df['Revenue']

# Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions for the next 5 years
future_years = np.arange(2024, 2029).reshape(-1, 1)
predictions = model.predict(future_years)

# Add predictions to the DataFrame
future_df = pd.DataFrame({'Year': future_years.flatten(), 'Revenue': predictions})

# Plot the forecast
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Revenue'], marker='o', label='Historical Revenue')
plt.plot(future_df['Year'], future_df['Revenue'], marker='x', linestyle='--', label='Forecasted Revenue')
plt.xlabel('Year')
plt.ylabel('Revenue (in billions USD)')
plt.title('Market Revenue Forecast')
plt.legend()
plt.show()
```

```python
# Prepare the data for the regression model
X = df[['Year']]
y = df['Revenue']

# Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions for the next 5 years
future_years = np.arange(2024, 2029).reshape(-1, 1)
predictions = model.predict(future_years)

# Add predictions to the DataFrame
future_df = pd.DataFrame({'Year': future_years.flatten(), 'Revenue': predictions})

# Plot the forecast
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Revenue'], marker='o', label='Historical Revenue')
plt.plot(future_df['Year'], future_df['Revenue'], marker='x', linestyle='--', label='Forecasted Revenue')
plt.xlabel('Year')
plt.ylabel('Revenue (in billions USD)')
plt.title('Market Revenue Forecast')
plt.legend()
plt.show()

# Financial Model Parameters
pricing_per_user = 100  # USD
fixed_costs = 10  # million USD

# Calculate total profit for future years
future_df['Total Sales'] = future_df['Revenue']
future_df['Total Profit'] = pricing_per_user * future_df['Total Sales'] - fixed_costs

# Display the financial forecast
print(future_df[['Year', 'Revenue', 'Total Profit']])
```
